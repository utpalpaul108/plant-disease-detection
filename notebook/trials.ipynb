#### Experiment With ensure_annotations

```python
from ensure import ensure_annotations
```

```python
def get_products(x:int, y:int)->int:
    return x*y
```

```python
get_products(2,'5')
```

```python
@ensure_annotations
def get_products(x:int, y:int)->int:
    return x*y
```

```python
get_products(2,'5')
```
